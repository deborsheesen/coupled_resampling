### Consider a few representative jumps of MCMC to tune the number of particles


* We want the variance of the delta log-likelihood to be of order 1.

```python
# Calling libraries:
from __future__ import division
%matplotlib inline
import numpy as np
import time
from scipy.stats import norm, uniform, bernoulli
from scipy.linalg import sqrtm 
from pylab import plot, show, legend
import matplotlib.pyplot as plt
import math 
import scipy
from model_functions import g, likelihood_PF_proposed, likelihood_cpm
int = np.vectorize(int) 
```

* Specify parameters and load data.

```python
n = 3

#setup measurement process
observation_points = np.arange(0,n)
H = np.zeros(( n, n ))
for i in range(n): H[ i, observation_points[i] ] = 1

observation_noise = 0.3
R = observation_noise*np.diag(np.ones(n))
```

### Define a few functions:

```python
def representative_variances(theta_0, y, x_0, N_particles, rep, n_jumps, sigma_theta, g, H, R,
                             rho_cpm, dt, delta, clambda=50, n_Sinkhorn=50, q=50, uv_threshold=10**10, 
                             print_jump_index=1) :

    scipy.random.seed()
    np.random.seed()
    
    n = len(x_0)
    theta_dim = len(theta_0)
    
    theta_values      = np.zeros((theta_dim, 2) ) 
    theta_values[:,0] = theta_0

    variances = np.zeros(( n_jumps, 2 ))
    run_times = np.zeros(2)

    delta_loglikelihood = np.zeros(( rep, 2 ))

    start_time = time.clock()
    for n in range(n_jumps) : 
        
        scipy.random.seed()
        theta_values[0:3,1] = theta_values[0:3,0] + sigma_theta[0:3]*np.random.randn(theta_dim-1)
        theta_values[-1 ,1] = np.exp( np.log(theta_values[-1,0]) + sigma_theta[-1]*np.random.randn(1) ) 
        
        for r in range(rep) :
            L_cpm = likelihood_cpm(N_particles[0], theta_values, y, x_0, rho_cpm, dt, delta, g, H, R)
            L_proposed = likelihood_PF_proposed(N_particles[1],theta_values, y, x_0, dt, delta, g, H, R, 
                                                'OT', clambda, n_Sinkhorn, q, uv_threshold)
            delta_loglikelihood[r,0] = L_cpm[1] - L_cpm[0]
            delta_loglikelihood[r,1] = L_proposed[1] - L_proposed[0]
        
        variances[n,0] = np.var(delta_loglikelihood[:,0])        
        variances[n,1] = np.var(delta_loglikelihood[:,1])
        
        if (n+1)%print_jump_index == 0 :
            print('Jump', n+1, 'over in', round((time.clock()-start_time)/60,1), 'minutes')
            
    return variances, delta_loglikelihood
```

```python
def plot_jump_variances(theta, y, x_0, N_particles, rep, n_jumps, sigma_theta, g, 
                        H, R, rho_cpm, dt, delta, clambda=50, n_Sinkhorn=50, q=50, uv_threshold=10**10,
                        print_jump_index=1) :
    
    var, delta_loglikelihood = representative_variances(theta, y, x_0, N_particles, rep,
                                                        n_jumps, sigma_theta, g, H, R, rho_cpm, dt, delta,
                                                        clambda, n_Sinkhorn, q, uv_threshold,
                                                        print_jump_index)

    plt.rcParams['figure.figsize'] = (14.0, 4.0)
    plt.plot(var[:,0],'bo-',label='correlated pseudo-marginal')
    plt.plot(var[:,1],'ro-',label='optimal transport')
    plt.title('Variance of representative jumps of MCMC',fontsize=14)
    plt.xlabel('Jump number',fontsize=12)
    plt.ylabel('Variance',fontsize=12)
    plt.grid(True)
    plt.legend(frameon=False)
    plt.show()
    
    print ('Mean variance of CPM and OT:', np.mean(var,axis=0))
    print ('Ratio of CPM over OT       :', np.mean(var,axis=0)[0]/np.mean(var,axis=0)[1])
    
    return var, delta_loglikelihood
```
