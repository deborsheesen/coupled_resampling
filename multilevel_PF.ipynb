```python
from __future__ import division
%matplotlib inline
import numpy as np
import time
from scipy.stats import norm, uniform, bernoulli, gaussian_kde
from scipy.linalg import sqrtm 
from pylab import plot, show, legend
import matplotlib.pyplot as plt
import math 
import scipy
from model_functions import g, MLPF
tnrfont = {'fontname':'Times New Roman'}
from matplotlib import gridspec
import matplotlib as mpl
mpl.rc('font',family='Times New Roman')
plt.rcParams['figure.figsize'] = (18.0, 5.0)
int = np.vectorize(int) 
```

```python
dt    = 10**-3 / 5
delta = 10*dt
T     = 1000
n     = 3

sigma, rho, beta, noise_intensity = 10, 28, 8/3, 0.1
theta = [sigma, rho, beta, noise_intensity]

#setup measurement process

space_between_observations = 1

observation_points = np.arange(0,n)
m = len(observation_points)
H = np.zeros(( m, n ))
for i in range(m):
    H[ i, observation_points[i] ] = 1

sigma_error = 0.3    
R = sigma_error*np.diag(np.ones(m))

y = np.loadtxt('y_e3.out'); x_0 = np.loadtxt('x_0_e3.out'); latent = np.loadtxt('latent_e3.out')
```

```python
def test_function(x) :
    return np.sum(x,axis=0)
```

```python
n_particles = 100
rep = 1000

delta_logNC  = np.zeros((2, T, rep))
test_fn_val  = np.zeros((2, T, rep))
running_time = np.zeros((2, T, rep)) 

for r in range(rep) :
    
    maximal_mlpf = MLPF(n_particles, theta, y, x_0, dt, delta, g, H, R, test_function,'maximal')
    delta_logNC[0,:,r] = maximal_mlpf[0][0,:] - maximal_mlpf[0][1,:]
    test_fn_val[0,:,r] = maximal_mlpf[1][0,:] - maximal_mlpf[1][1,:]
    running_time[0,:,r] = maximal_mlpf[2]
    
    OT_mlpf = MLPF(n_particles, theta, y, x_0, dt, delta, g, H, R, test_function,
                   'OT', clambda=100, n_Sinkhorn=150, q=50, uv_threshold=10**30)
    delta_logNC[1,:,r] = OT_mlpf[0][0,:] - OT_mlpf[0][1,:]
    test_fn_val[1,:,r] = OT_mlpf[1][0,:] - OT_mlpf[1][1,:]
    running_time[1,:,r] = OT_mlpf[2]
```

```python
delta_logNC_performance = np.zeros(( 2, 1000, 25 ))
test_fn_val_performance = np.zeros(( 2, 1000, 25 ))

for r in range(25) :
    delta_logNC_performance[:,:,r] = (np.var(delta_logNC[:,:,r*40:(r+1)*40],axis=2)*
                                      np.mean(running_time[:,:,r*40:(r+1)*40],axis=2))
    test_fn_val_performance[:,:,r] = (np.var(test_fn_val[:,:,r*40:(r+1)*40],axis=2)*
                                      np.mean(running_time[:,:,r*40:(r+1)*40],axis=2))
    
relative_performance_delta_logNC = delta_logNC_performance[0,:,:]/delta_logNC_performance[1,:,:] 
relative_performance_test_fn_val = test_fn_val_performance[0,:,:]/test_fn_val_performance[1,:,:]    
```

```python
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16)     
fig = plt.figure(1)  
    
x_vals = np.arange(1,1001)  
ax = fig.add_subplot(121)
ax.plot(x_vals, np.median(relative_performance_delta_logNC,axis=1),'b-',markersize=8)
plt.plot(x_vals, np.percentile(relative_performance_delta_logNC,axis=1,q=5), 'r-', alpha=0.75, markersize=8)
plt.plot(x_vals,np.percentile(relative_performance_delta_logNC,axis=1,q=95),'r-', alpha=0.75, markersize=8)
plt.ylabel('Relative efficiency',fontsize=17,**tnrfont)
plt.yscale('log')
plt.grid(True) 
plt.ylim(2*10**(-1),2*10**2)

ax = fig.add_subplot(122)
ax.plot(x_vals, np.median(relative_performance_test_fn_val,axis=1),'b-',markersize=8,label='median')
plt.plot(x_vals, np.percentile(relative_performance_test_fn_val,axis=1,q=5), 
         'r-',label='90% confidence band', alpha=0.75, markersize=8)
plt.plot(x_vals,np.percentile(relative_performance_test_fn_val,axis=1,q=95),'r-', alpha=0.75, markersize=8)
plt.yscale('log')
plt.grid(True)
plt.ylim(2*10**(-1),2*10**2)

fig.text(0.724, 0.93, r'Estimating $\Delta l(\theta \mid y_0, \ldots, y_{k \delta})$', 
         ha='center',fontsize=19,**tnrfont)
fig.text(0.3, 0.93, r'Estimating $\mathsf{E} [ \phi(X_{k \delta}) | y_0, \ldots, y_{k \delta}]$', 
         ha='center',fontsize=19,**tnrfont)

fig.text(0.5, 0, 'Time index', ha='center', fontsize = 17, **tnrfont)
ax.legend(bbox_to_anchor=(0.3, 1.2), frameon = False, fontsize=16, ncol = 2)
```
